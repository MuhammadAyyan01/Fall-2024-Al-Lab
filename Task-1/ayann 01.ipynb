```python


tasks = []  

def add_new_task():
    task = input("Enter new task: ")
    tasks.append({"task": task, "done": False})
    print("Task added!")

def show_tasks():
    print("\n--- Your Current Tasks ---")
    if not tasks:
        print("No tasks available.")
    else:
        counter = 1
        for task_item in tasks:
            status_text = "Done" if task_item["done"] else "Not Done"
            print(f"{counter}. {task_item['task']} [{status_text}]")
            counter += 1

def mark_task_complete():
    show_tasks()
    if len(tasks) > 0:
        try:
            user_input = input("Which task number did you complete? ")
            task_number = int(user_input)
            index = task_number - 1
            tasks[index]["done"] = True
            print("Task marked as completed!")
        except ValueError:
            print("Please enter a valid number!")
        except IndexError:
            print("That task number doesn't exist!")

def delete_task():
    show_tasks()
    if len(tasks) > 0:
        try:
            task_to_remove = int(input("Enter the task number to delete: "))
            removed_item = tasks[task_to_remove - 1]
            tasks.pop(task_to_remove - 1)
            print(f"Deleted: '{removed_item['task']}' from your list")
        except (IndexError, ValueError):
            print("That doesn't look like a valid task number!")


while running:
    print("\n===== TO-DO LIST MENU =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Done")
    print("4. Delete Task")
    print("5. Exit")

    user_choice = input("What would you like to do? (1-5): ")

    if user_choice == "1":
        add_new_task()
    elif user_choice == "2":
        show_tasks()
    elif user_choice == "3":
        mark_task_complete()
    elif user_choice == "4":
        delete_task()
    elif user_choice == "5":
        print("Thanks for using the to-do list! See you later!")
        running = False
    else:
        print("Please pick a number between 1 and 5.")
```
